#### My understanding of the assignment 

We define a start_value or initial value - x_0

For each time period 't' -> t = 0, 1, 2, 3,...., n:

x_t = x_t-1 + e_t, where e_t is a random number generated from standard normal distribution [e_t ~ N(0,1)]

We need to then record these values for given 'x_0' and 'n in a series. 

Next, we sort the values x_t in ascending order and divide these into four groups/quartiles/bins which represent lowest 25% values, next 25% values, another next 25% values and highest 25% values.

```python
import numpy as np
import pandas as pd
```

```python
#I want to generate a time-series with 'n' time steps starting from start_value. At each step, I want to
#add one random shock from a standard normal distribution (mean = 0, variance = 1) and append it to the list. 

#At this point, I am just setting up the program and not allocating any specific values to n and start_value. 

def my_series(n,start_value):
    values = []                                #Create an empty list of values to store 
    current = float(start_value)               #Defined a new variable 'current' to ensure decimal values can be included in the list
    t = 0                                      #This is the initial time step, alternately, I am defining my first 't' 
    while t < n:
        for i in range(1):
            shock = np.random.randn()
            current = current + shock
            values.append(current)
            t = t + 1
    return pd.Series(values, name = "my_series") 

#Example 1 - applying the program to one scenario where n = 30 and start_value = 100 (as indicated in the Problem Set)
n = 30                                       #Number of time steps
start_value = 100                            #Initial value 
s = my_series(n, start_value)                #Generate a series as defined

bins = pd.qcut(s, q=4)                       #Divide the Series into 4 quantile-based bins (quartiles)
print(bins)  
```
